```python
import pandas as pd
import numpy as np

df=pd.read_excel(r"C:\Users\liman\Documents\Main Academy\SpongeBobs\datasets for corals\corals_dataset.xlsx")
df
```

```python
import re

# Function to extract floats without degrees
def extract_float_without_degree(value):
    if isinstance(value, str):
        match = re.search(r'([\d.]+)', value)
        if match:
            return match.group(1)
    return value

# Apply the function to the column
df['January_temp'] = df['January_temp'].apply(extract_float_without_degree)

# Convert the column to float data type
df['January_temp'] = pd.to_numeric(df['January_temp'], errors='coerce')

# Apply the function to the column
df['June_temp'] = df['June_temp'].apply(extract_float_without_degree)

# Convert the column to float data type
df['June_temp'] = pd.to_numeric(df['June_temp'], errors='coerce')

df

```

```python
df.to_excel('corals_dataset.xlsx', index=False, index_label=False)
```

```python
df.to_csv('corals_dataset.csv', index=False, index_label=False)
```
