```python
import xarray as xr
import pandas as pd
```

```python
# Select a subset of the data January 2022
ds = xr.open_dataset(r'C:\Users\liman\Documents\Main Academy\SpongeBobs\cmems_obs-sst_glo_phy_nrt_l4_P1D-m_1683646929672.nc')
# convert the xarray DataArray to a Pandas DataFrame
df = ds.to_dataframe()

# reset the index to create separate columns for lat, lon and time
df = df.reset_index()
df
```

```python
df['lat'] = df['lat'].apply(lambda x: int(round(x)))
df
```

```python
df['lon'] = df['lon'].apply(lambda x: int(round(x)))
df
```

```python
seven_seas_lat_lon=pd.read_excel('seven seas lat lon.xlsx')
seven_seas_lat_lon
```

```python
no_nan_temp = ~df['analysed_sst'].isna()

# this will return a boolean Series with True for non-NaN values and False for NaN values
# you can use this Series to index into your DataFrame and display only the rows where "lat" has no NaN values:
df_no_nan_temp = df[no_nan_temp]
print(df_no_nan_temp)
```

```python
merged_df_2 = pd.merge(df_no_nan_temp, seven_seas_lat_lon, on=['lat','lon'])
merged_df_2
```

```python
merged_df_2 = merged_df_2.drop(['Unnamed: 0.2', 'Unnamed: 0.1', 'Unnamed: 0'],  axis=1)
merged_df_2
```

```python
#Convert to Celsius from Kelvin
merged_df_2['analysed_sst'] = merged_df_2['analysed_sst'] - 273.15
merged_df_2
```

```python
#the main file with t for work total temperature
merged_df_2.to_excel('seven seas temperature 2022 Jan total.xlsx')
```

```python
merged_df_2.to_csv('seven seas temperature 2022 Jan total.csv')
```

```python
# Calculate the mean of 'analysed_sst' for each timestamp and sea
mean_sst = merged_df_2.groupby(['time', 'name'])['analysed_sst'].mean().reset_index()
mean_sst
```

```python
#the main file with t for work
mean_sst.to_excel('seven seas temperature 2022 Jan.xlsx')
```
