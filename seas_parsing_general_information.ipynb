```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import requests
import json

#parsing the Largest seas with areas
seas_list = []

driver_path = 'C:/Users/liman/Desktop/chromedriver_win32/chromedriver.exe'
options = webdriver.ChromeOptions()
options.add_argument('--headless')
driver = webdriver.Chrome(driver_path, options=options)

url = "https://en.wikipedia.org/wiki/List_of_seas"
driver.get(url)
driver.implicitly_wait(1)
soup = BeautifulSoup(driver.page_source, 'html.parser')

seas_div = soup.find('div', class_='div-col')  # select the div with the seas data
seas_ol = seas_div.find('ol')  # select the ol inside the div
seas_items = seas_ol.find_all('li')  # select all the li elements inside the ol

for item in seas_items:
    seas_dict = {}
    sea_name = item.find('a').text
    sea_link = item.find('a')['href']
    sea_area = item.find('sup').previous_sibling.strip().replace(',', '')  # extract the area from the text and clean it up
        
    seas_dict["name"] = sea_name
    seas_dict["link"] = sea_link
    seas_dict["area"] = sea_area
    
#     seas_list.append(seas_dict)
    
    url_2 = "https://en.wikipedia.org" + sea_link
    driver.get(url_2)
    driver.implicitly_wait(1)
    soup_2 = BeautifulSoup(driver.page_source, 'html.parser')
    try: 
        latitude = soup_2.find('span', {'class': 'latitude'}).text
        longitude = soup_2.find('span', {'class': 'longitude'}).text
    except AttributeError:
        latitude = "Not found"
        longitude = "Not found"
        
    seas_dict["latitude"] = latitude
    seas_dict["longitude"] = longitude
    
  
        
    seas_list.append(seas_dict)
    
    
# 
print(seas_list)

driver.quit()
```

```python
import pandas as pd

seas = pd.read_excel('largest seas long lat.xlsx')
seas
```

```python
import pandas as pd
import re

df=pd.read_excel('seven seas.xlsx')
df
```

```python
def convert_latitude(val):
    if val.endswith('N'):
        return float(val[:-2])
    elif val.endswith('S'):
        return -1 * float(val[:-2])
    else:
        return None # or raise an exception if you want to handle invalid values

df['lat'] = df['latitude'].apply(convert_latitude)
df
```

```python
def convert_longitude(val):
    if val.endswith('E'):
        return float(val[:-2])
    elif val.endswith('W'):
        return -1 * float(val[:-2])
    else:
        return None # or raise an exception if you want to handle invalid values
```

```python
df['lon'] = df['longitude'].apply(convert_longitude)
df
```

```python
df = df.drop(['Unnamed: 0.1', 'Unnamed: 0'],  axis=1)
```

```python
df.to_csv('seven seas lat lon.csv', index=False, index_label=False)
df.to_excel('seven seas lat lon.xlsx', index=False, index_label=False)
```
